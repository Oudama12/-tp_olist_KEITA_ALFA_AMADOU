## Valeurs manquantes

```python
def missing_report(df, name):
    pct = round(df.isna().sum() / len(df) * 100, 2)
    pct = pct[pct > 0].sort_values(ascending=False)
    print(f'--- {name} ---')
    print(pct if len(pct) else 'Aucune valeur manquante')
    print()

for name, df in [('orders',orders),('items',items),('products',products),
                 ('customers',customers),('reviews',reviews),
                 ('sellers',sellers),('payments',payments),('categories',categories)]:
    missing_report(df, name)
```

# Partie 2 - Nettoyage, Feature Engineering et Fusion

## Suppression des colonnes inutiles

```python
# Supprimons les colonnes qui ne servent pas pour l'analyse
items = items.drop(columns=['order_item_id', 'shipping_limit_date'])

products = products.drop(columns=['product_name_lenght','product_description_lenght',
                                  'product_photos_qty','product_length_cm',
                                  'product_height_cm','product_width_cm'])

customers = customers.drop(columns=['customer_unique_id','customer_zip_code_prefix'])

reviews = reviews[['order_id','review_score']]

sellers = sellers.drop(columns=['seller_zip_code_prefix'])

payments = payments.drop(columns=['payment_sequential'])

print('Colonnes inutiles supprimees')
```

## Conversion des types temporels

```python
date_cols = ['order_purchase_timestamp',
             'order_approved_at',
             'order_delivered_customer_date',
             'order_estimated_delivery_date']

orders[date_cols] = orders[date_cols].apply(pd.to_datetime)
orders.dtypes[date_cols]
```

## Calcul des delais de livraison

```python
orders['delivery_days'] = (
    orders['order_delivered_customer_date']
    - orders['order_purchase_timestamp']
).dt.days

orders['delay_days'] = (
    orders['order_delivered_customer_date']
    - orders['order_estimated_delivery_date']
).dt.days

orders[['delivery_days','delay_days']].describe()
```

## Question 5 - Mediane et taux de retard

```python
mediane = orders['delivery_days'].median()
pct_retard = (orders['delay_days'] > 0).mean() * 100

print(f'Mediane du delai de livraison : {mediane} jours')
print(f'Pourcentage de commandes en retard : {pct_retard:.2f} %')
```

# Remarque
La mediane du delai de livraison est d'environ 10 jours et environ 7 a 8 % des commandes sont livrees en retard. La majorite des commandes arrivent donc avant la date estimee.

## Fusion des tables

```python
# Etape 1 : commandes + articles
df = orders.merge(items, on='order_id', how='inner')
print('Apres orders+items :', df.shape)
```

```python
# Etape 2 : + produits avec traduction de categorie
products = products.merge(categories, on='product_category_name', how='left')
df = df.merge(products[['product_id','product_category_name_english','product_weight_g']],
              on='product_id', how='inner')
print('Apres +products :', df.shape)
```

```python
# Etape 3 : + vendeurs
df = df.merge(sellers[['seller_id','seller_state']], on='seller_id', how='inner')
print('Apres +sellers :', df.shape)
```

```python
# Etape 4 : + clients
df = df.merge(customers[['customer_id','customer_state','customer_city']],
              on='customer_id', how='left')
print('Apres +customers :', df.shape)
```

```python
# Etape 5 : + notes de satisfaction
df = df.merge(reviews, on='order_id', how='left')
print('Apres +reviews :', df.shape)
```

# Partie 3 - Analyse Descriptive et Agregations

## Evolution du chiffre d'affaires mensuel

```python
df['month'] = df['order_purchase_timestamp'].dt.to_period('M').astype(str)

ca_mensuel = (
    df.groupby('month')
      .agg(ca_total=('price','sum'), nb_commandes=('order_id','nunique'))
      .reset_index()
)
ca_mensuel.head()
```

## Question 8 - Graphique du CA mensuel

```python
fig, ax1 = plt.subplots(figsize=(13,5))

ax1.plot(ca_mensuel['month'], ca_mensuel['ca_total'], color='steelblue', marker='o', label='CA total (BRL)')
ax1.set_xlabel('Mois')
ax1.set_ylabel('Chiffre d\'affaires (BRL)', color='steelblue')
ax1.tick_params(axis='x', rotation=45)

ax2 = ax1.twinx()
ax2.plot(ca_mensuel['month'], ca_mensuel['nb_commandes'], color='orange', marker='s', label='Nb commandes')
ax2.set_ylabel('Nombre de commandes', color='orange')

plt.title('Evolution du chiffre d\'affaires et du nombre de commandes par mois')
fig.tight_layout()
plt.savefig('../visuals/ca_mensuel.png', dpi=150)
plt.show()
```

# Remarque
On observe une forte croissance du CA et du nombre de commandes entre fin 2016 et fin 2017. Un pic tres visible apparait en novembre 2017 qui correspond au Black Friday bresilien. Les premiers mois (2016) sont presque vides : la plateforme demarrait. Apres le pic de fin 2017, l'activite se stabilise en 2018.

## Top categories de produits

## Delai de livraison par etat

```python
delai_etat = (
    df.groupby('customer_state')['delivery_days']
      .mean()
      .sort_values(ascending=False)
      .reset_index()
      .rename(columns={'delivery_days':'delai_moyen_j'})
)
print('Etats les plus lents :')
print(delai_etat.head(5))
print('\nEtats les plus rapides :')
print(delai_etat.tail(5))
```

## Question 10 - Barplot des etats extremes

```python
extremes = pd.concat([delai_etat.head(5), delai_etat.tail(5)])

plt.figure(figsize=(10,6))
sns.barplot(data=extremes, y='customer_state', x='delai_moyen_j',
            hue='customer_state', palette='coolwarm', legend=False)
plt.title('Etats avec les delais de livraison les plus longs et les plus courts')
plt.xlabel('Delai moyen (jours)')
plt.ylabel('Etat')
plt.tight_layout()
plt.savefig('../visuals/delai_par_etat.png', dpi=150)
plt.show()
```

# Remarque
L'etat le plus lent est **Roraima (RR)** avec environ 28 jours de delai moyen, et le plus rapide est **Sao Paulo (SP)** avec environ 8 jours. L'explication est geographique : Roraima est tout au nord du Bresil, en pleine Amazonie, loin des grands centres logistiques. Sao Paulo concentre la plupart des vendeurs et entrepots, donc les commandes y sont livrees beaucoup plus vite.

## Satisfaction client et delai de livraison

```python
bins = [0, 7, 14, 21, 999]
labels = ['0-7j', '7-14j', '14-21j', '21j+']
df['delivery_bucket'] = pd.cut(df['delivery_days'], bins=bins, labels=labels)

satisfaction = (
    df.groupby('delivery_bucket', observed=True)['review_score']
      .mean()
      .reset_index()
)
satisfaction
```

## Question 11 - Barplot satisfaction par tranche de delai

```python
plt.figure(figsize=(8,5))
sns.barplot(data=satisfaction, x='delivery_bucket', y='review_score',
            hue='delivery_bucket', palette='RdYlGn_r', legend=False)
plt.title('Note moyenne du client par tranche de delai de livraison')
plt.xlabel('Tranche de delai')
plt.ylabel('Note moyenne (1-5)')
plt.ylim(0, 5)
plt.tight_layout()
plt.savefig('../visuals/satisfaction_vs_delai.png', dpi=150)
plt.show()
```

# Remarque
Il y a une correlation tres claire : plus le delai de livraison est long, plus la note baisse. Les commandes livrees en moins de 7 jours obtiennent en moyenne 4.3/5, tandis que celles livrees en plus de 21 jours descendent autour de 2.7/5. Le delai de livraison est donc un facteur cle de la satisfaction client.

## Exercice 3 - Analyse libre : mode de paiement et panier moyen

```python
# On rejoute la table payments pour analyser les paiements
pay = pd.read_csv(BASE + 'olist_order_payments_dataset.csv')

pay_analyse = (
    pay.groupby('payment_type')
       .agg(nb=('order_id','count'), panier_moyen=('payment_value','mean'))
       .sort_values('nb', ascending=False)
       .reset_index()
)
pay_analyse
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13,5))

sns.barplot(data=pay_analyse, x='payment_type', y='nb', ax=axes[0],
            hue='payment_type', palette='Blues_r', legend=False)
axes[0].set_title('Nombre de paiements par mode')
axes[0].set_xlabel('Mode de paiement')
axes[0].set_ylabel('Nombre de transactions')

sns.barplot(data=pay_analyse, x='payment_type', y='panier_moyen', ax=axes[1],
            hue='payment_type', palette='Greens_r', legend=False)
axes[1].set_title('Panier moyen par mode de paiement')
axes[1].set_xlabel('Mode de paiement')
axes[1].set_ylabel('Panier moyen (BRL)')

plt.tight_layout()
plt.savefig('../visuals/analyse_paiements.png', dpi=150)
plt.show()
```

# Remarque
La carte de credit est de loin le mode de paiement le plus utilise (environ 75 % des transactions). Le boleto (virement bancaire bresilien) arrive en second. Le panier moyen est tres proche entre les modes de paiement, autour de 150-170 BRL, avec un panier legerement plus eleve pour la carte de credit grace au paiement en plusieurs fois.

## Question 14 - Conclusion et recommandations

**Principaux enseignements de l'analyse :**

1. **Le delai de livraison est le facteur cle de la satisfaction client** : les commandes livrees en plus de 21 jours obtiennent une note moyenne autour de 2.7/5, contre 4.3/5 pour celles livrees en moins de 7 jours.
2. **Il existe une forte inegalite geographique** : les etats du nord (Roraima, Amapa) attendent jusqu'a 28 jours en moyenne alors que Sao Paulo recoit ses commandes en 8 jours. La logistique est concentree dans le sud-est du pays.
3. **Le marche Olist est domine par quelques categories** : health_beauty, watches_gifts et bed_bath_table representent une part importante du CA, et le pic d'activite a lieu pendant le Black Friday de novembre.

**Recommandations pour l'equipe produit Olist :**

- **Ouvrir des entrepots regionaux dans le nord du Bresil** (par exemple a Manaus) pour reduire le delai de livraison dans les etats les plus eloignes et donc augmenter la satisfaction client dans ces zones.
- **Mettre en place une garantie de livraison rapide** pour les categories a forte marge (watches_gifts, computers_accessories) afin de fideliser les clients et de profiter du pic du Black Friday avec une logistique mieux dimensionnee.
